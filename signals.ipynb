## HeaviSine

$f(t) = 4 \sin(4 \pi t) - sgn (t - 0.3) - sgn (0.72 - t)$

```python
N            = 2048
t, heavisine = generateHeaviSine(N)

plt.plot(t, heavisine, "-k", linewidth = 1)
plt.xlim(0,1)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("t", fontsize = 22)
plt.ylabel("f(t)", fontsize = 22)
```

## Blocks

$f(t) = \sum \; h_{j} \, K(t - t_{j})$

$K(t) = \dfrac{1 \, + \, sgn(t)}{2}$

$pos = [0.1, \, 0.13, \, 0.15, \, 0.23, \, 0.25, \, 0.40, \, 0.44, \, 0.65, \, 0.76, \, 0.78, \, 0.81]$

$hgt = [4.0, \, -5.0, \, 3.0, \, -4.0, \, 5.0, \, -4.2, \, 2.1, \, 4.3, \, -3.1, \, 2.1, \, -4,2]$

```python
N         = 2048
t, blocks = generateBlocks(N)

plt.figure(num = None, figsize=(8,6), dpi=100)
plt.plot(t, blocks, "-k", linewidth = 1)
plt.xlim(0,1)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("t", fontsize = 22)
plt.ylabel("f(t)", fontsize = 22)
plt.show()
```

## Bumps

$f(t) = \sum \; h_{j} \, K \left( \dfrac{t-t_{j}}{w_{j}} \right)$

$K(t) = \left( \, 1 \, + \, | \, t \, | \, \right)^{-4}$

$pos = [0.10, \, 0.13, \, 0.15, \, 0.23, \, 0.25, \, 0.40, \, 0.44, \, 0.65, \, 0.76, \, 0.78, \, 0.81]$

$hgt = [4.0, \, 5.0, \, 3.0, \, 4.0, \, 5.0, \, 4.2, \, 2.1, \, 4.3, \, 3.1, \, 5.1, \, 4.2]$

$wgt = [0.005, \, 0.005, \, 0.006, \, 0.010, \, 0.010, \, 0.030, \, 0.010, \, 0.010, \, 0.005, \, 0.008, \, 0,005]$

```python
N        = 2048
t, bumps = generateBumps(N)

plt.figure(num = None, figsize=(8,6), dpi=100)
plt.plot(t, bumps, "-k", linewidth = 1)
plt.xlim(0,1)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("t", fontsize = 22)
plt.ylabel("f(t)", fontsize = 22)
plt.show()
```

## Doppler 

$f(t) = \sqrt{t \, (1 \, - \, t)} \; \sin \left( \dfrac{2 \, \pi \, (1 \, + \, \varepsilon)}{t \, + \, \varepsilon} \right)$

$\varepsilon = 0,05$

```python
N          = 2048
t, doppler = generateDoppler(N, 0.05)

plt.figure(num = None, figsize=(8,6), dpi=100)
plt.plot(t, doppler, "-k", linewidth = 1)
plt.xlim(0,1)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel("t", fontsize = 22)
plt.ylabel("f(t)", fontsize = 22)
plt.show()
```
